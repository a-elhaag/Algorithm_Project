```python
test_array = {
    "sorted": [1, 2, 3, 4, 5],
    "reverse_sorted": [5, 4, 3, 2, 1],
    "mixed": [3, 1, 4, 2, 5],
    "duplicates": [1, 2, 2, 3, 4],
    "empty": [],
    "single_element": [1],
}
```

```python
def insertion_sort(arr):
    # Traverse from the second element to the end
    for i in range(1, len(arr)):
        key = arr[i]  # Current element to insert
        j = i - 1
        # Shift elements that are greater than key to the right
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        # Place the key in its correct sorted position
        arr[j + 1] = key
    return arr
```

```python
print("Insertion Sort Test Cases:")
for name, arr in test_array.items():
    sorted_arr = insertion_sort(arr.copy())
    print(f"{name}: {sorted_arr}")
    assert sorted_arr == sorted(arr), f"Test failed for {name}"
```

```python
import random

def quick_sort(arr):
    # Base case: array of length 0 or 1 is already sorted
    if len(arr) <= 1:
        return arr

    # Randomly pick a pivot element
    pivot = random.choice(arr)

    # Partition the array into three parts:
    # Elements less than, equal to, and greater than the pivot
    less = [x for x in arr if x < pivot]
    equal = [x for x in arr if x == pivot]
    greater = [x for x in arr if x > pivot]

    # Recursively sort the less and greater partitions and concatenate the result
    return quick_sort(less) + equal + quick_sort(greater)
```

```python
print("Quick Sort Test Cases:")
for name, arr in test_array.items():
    sorted_arr = quick_sort(arr.copy())
    print(f"{name}: {sorted_arr}")
    assert sorted_arr == sorted(arr), f"Test failed for {name}"
```

```python
def merge_sort(arr):
    # Base case: a list with 0 or 1 element is already sorted
    if len(arr) <= 1:
        return arr

    # Divide the array into two halves
    mid = len(arr) // 2
    left = merge_sort(arr[:mid])
    right = merge_sort(arr[mid:])

    # Merge the two sorted halves
    merged = []
    i = j = 0

    # Compare elements from left and right arrays one by one
    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            merged.append(left[i])
            i += 1
        else:
            merged.append(right[j])
            j += 1

    # Append any remaining elements from either side
    merged.extend(left[i:])
    merged.extend(right[j:])

    return merged
```

```python
print("Merge Sort Test Cases:")
for name, arr in test_array.items():
    sorted_arr = merge_sort(arr.copy())
    print(f"{name}: {sorted_arr}")
    assert sorted_arr == sorted(arr), f"Test failed for {name}"
```

```python
def generate_input(size, order="random"):
    arr = list(range(size))
    if order == "random":
        random.shuffle(arr)
    elif order == "reversed":
        arr = arr[::-1]
    return arr
```

```python
def show_sort_snapshots(sort_func, algorithm_name):
    sizes = [random.randint(5, 15) for _ in range(3)]
    input_types = ["sorted", "random", "reversed"]

    for input_type in input_types:
        arr = generate_input(sizes[0], input_type)
        print(f"\n{algorithm_name} | Input Type: {input_type.capitalize()}")
        print("Before:", arr)
        result = sort_func(arr.copy())
        print("After :", result)
```

```python
show_sort_snapshots(insertion_sort, "Insertion Sort")
```

```python
show_sort_snapshots(quick_sort, "Quick Sort")
```

```python
show_sort_snapshots(merge_sort, "Merge Sort")
```

```python
import time

def timed_sort(sort_func, arr):
    start = time.time()
    result = sort_func(arr.copy())
    end = time.time()
    return result, end - start
```

```python
import matplotlib.pyplot as plt
from ipywidgets import interact, IntSlider, Dropdown
import numpy as np
```

```python
def test_sorts(input_size, input_type):
    algorithms = {
        "Insertion Sort": insertion_sort,
        "Merge Sort": merge_sort,
        "Quick Sort": quick_sort,
    }

    input_array = generate_input(input_size, input_type)
    times = []
    labels = []

    for name, func in algorithms.items():
        _, t = timed_sort(func, input_array.copy())  # always use copy
        times.append(t)
        labels.append(name)

    # Better-scaled bounds (not theoretical, but relative)
    lower_bound = min(times) * 0.8
    upper_bound = max(times) * 1.5

    # Plot
    plt.figure(figsize=(10, 5))
    bars = plt.bar(labels, times, color="skyblue")

    # Annotate time values
    for bar, t in zip(bars, times):
        plt.text(
            bar.get_x() + bar.get_width() / 2, t, f"{t:.5f}", ha="center", va="bottom"
        )

    plt.axhline(
        y=lower_bound, color="green", linestyle="--", label="Approx. Lower Bound"
    )
    plt.axhline(y=upper_bound, color="red", linestyle="--", label="Approx. Upper Bound")
    plt.title(
        f"Execution Time of Sorting Algorithms ({input_type} - {input_size} items)"
    )
    plt.ylabel("Time (seconds)")
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
interact(
    test_sorts,
    input_size=IntSlider(min=10, max=100, step=10, value=50),
    input_type=Dropdown(options=["random", "sorted", "reversed"]),
)
```

```python
def plot_cases(sort_func, algorithm_name):
    sizes = np.arange(50, 500, 100)  # Input sizes
    cases = {
        "Best Case (Sorted)": "sorted",
        "Average Case (Random)": "random",
        "Worst Case (Reversed)": "reversed",
    }

    plt.figure(figsize=(10, 5))

    for label, input_type in cases.items():
        times = []
        for size in sizes:
            arr = generate_input(size, input_type)
            start = time.time()
            sort_func(arr.copy())
            end = time.time()
            times.append(end - start)
        plt.plot(sizes, times, marker="o", label=label)

    plt.title(f"{algorithm_name} - Real Execution Times")
    plt.xlabel("Input Size")
    plt.ylabel("Execution Time (s)")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
plot_cases(insertion_sort, "Insertion Sort")
```

```python
plot_cases(quick_sort, "Quick Sort")
```

```python
plot_cases(merge_sort, "Merge Sort")
```

```python
import pandas as pd
```

```python
def generate_time_table():
    sizes = [100, 500, 1000]
    input_types = ["sorted", "random", "reversed"]
    algorithms = {
        "Insertion Sort": insertion_sort,
        "Merge Sort": merge_sort,
        "Quick Sort": quick_sort,
    }

    data = []

    for size in sizes:
        for input_type in input_types:
            arr = generate_input(size, input_type)
            for name, func in algorithms.items():
                _, t = timed_sort(func, arr.copy())
                data.append([name, size, input_type, round(t, 6)])

    df = pd.DataFrame(data, columns=["Algorithm", "Size", "Input Type", "Time (s)"])
    return df
```

```python
df = generate_time_table()

for input_type in ["sorted", "random", "reversed"]:
    print(f"\n## Execution Times on {input_type.capitalize()} Input")
    display(
        df[df["Input Type"] == input_type]
        .drop(columns="Input Type")
        .reset_index(drop=True)
    )
```
